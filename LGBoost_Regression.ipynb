```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.xls")
```

```python
dataset = pd.get_dummies(dataset, drop_first= True)
bol_col = dataset.select_dtypes(include='bool').columns
dataset[bol_col] = dataset[bol_col].astype(int)
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
independent = dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
```

```python
dependent = dataset[['charges']]
```

```python
# from sklearn.model_selection import train_test_split
# X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size=0.3, random_state=0)
```

```python
# pip install lightgbm
```

```python
from lightgbm import LGBMRegressor
regressor = LGBMRegressor(boosting_type= 'gbdt', learning_rate= 0.1, n_estimators=50, random_state=0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
result = r2_score(y_test, y_pred)
```

```python
result
```
